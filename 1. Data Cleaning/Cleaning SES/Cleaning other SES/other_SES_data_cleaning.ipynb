```python
import pandas as pd
import numpy as np

df=pd.read_csv("SES_data.csv")
df
```

```python
renames={"SE_A00002_001":"total_population","SE_A00002_002":"population_density",
                      "SE_A14006_001":"Median_household_income","SE_A01004_001":"Median_age",
                      "SE_A12001_001":"population_25_older","SE_A12001_002":"less_than_high_school",
                      "SE_A12001_003":"high_school_graduate","SE_A12001_004":"some_college","SE_A12001_005":"bachelors_degree",
                     "SE_A12001_006":"masters_degree","SE_A12001_007":"professional_degree","SE_A12001_008":"doctorate_degree"}

df=df.rename(columns=renames)

df
```

```python
df.columns
```

```python
col_names = ["Geo_FIPS", "Geo_QName"] + list(renames.values())
df1 = df[col_names]
```

```python
df1
```

```python
df1["percent_high_school_graduate"]=df1["high_school_graduate"]/df1["population_25_older"]
df1["percent_some_college_education_or_more"]=(df1['some_college']+df1['bachelors_degree']+df1['masters_degree']+df1[ 'professional_degree']+df1['doctorate_degree'])/df1["population_25_older"]
df1
```

```python
df1 = df1.drop(columns=['population_25_older', 'less_than_high_school',"high_school_graduate","some_college","bachelors_degree"
                       ,"masters_degree","professional_degree","doctorate_degree"])
df1
```

```python
df1.to_csv("census_data_2.csv")
```
