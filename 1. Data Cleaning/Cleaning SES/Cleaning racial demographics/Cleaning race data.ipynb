```python
# importing libraries
import pandas as pd
import tqdm.notebook as tq
# reading the data
df=pd.read_csv("R13277620_SL050.csv")
# converting columns to readable format
df=df.rename(columns = {'SE_A00001_001':'Total_Population','SE_B04001_004':'blacks',
                    'SE_B04001_005':'Native','SE_B04001_006':'Asian','SE_B04001_010':'Hispanic'})
## converting the 5 digit FIPS into characters
df["Geo_FIPS"]=[str(int(i)) for i in tq.tqdm(df.Geo_FIPS)]
## converting the county FIPS to character, including the 0 infront of counties with 4-digit FIPS (eg "1234" -> "01234")
df["Geo_FIPS"]=(pd.Series(df.Geo_FIPS)).str.pad(width=5,fillchar='0')
cols=['Total_Population','blacks','Native','Asian','Hispanic']
df[cols] = df[cols].apply(pd.to_numeric)
## Computing only respective racial compositions (in terms of proportion)
df["percent_black"]=df['blacks']/df['Total_Population']
df["percent_hispanic"]=df['Hispanic']/df['Total_Population']
df["percent_native"]=df['Native']/df['Total_Population']
df["percent_asian"]=df['Asian']/df['Total_Population']
# Selecting the columns and saving the data
df_new=df[['Geo_FIPS','Geo_QName','Total_Population',"percent_black","percent_hispanic","percent_native","percent_asian"]]
df_new.to_csv("race_data.csv",index=False)
```
